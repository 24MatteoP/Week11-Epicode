```python
# ESERCIZIO 
import numpy as np
import pandas as pd

years = 5
guadagni = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years), 
                         "Anno": np.repeat(list(range(years)), 12), 
                         "Valore": np.random.randint(800, 5000, 12*years)})
```

```python
guadagni
```

```python
# CALCOLA LA SOMMA CUMULATIVA DEI GUADAGNI UTILIZZANDO .CUMSUM()

guadagni['ValoreCumulativo'] = guadagni['Valore'].cumsum()
guadagni
```

```python
# COME SOPRA, MA RAGGRUPPATO PER OGNI ANNO USANDO PRIMA UN .GROUPBY

guadagni['ValoreCumulativo'] = guadagni.groupby('Anno')['Valore'].cumsum()
guadagni
```

```python
import os
import dotenv
import sqlalchemy
import pandas as pd

# Carico le variabili d'ambiente dal file cred.env
dotenv.load_dotenv(dotenv_path="cred.env", override=True)

# Recupero i valori delle variabili d'ambiente
username = os.getenv("USERNAME")  # Assicurati che nel file .env ci sia USERNAME=studente_dapt
password = os.getenv("PASSWORD")  # PASSWORD=Ep1c0d3!!D4t4**4n4lys1s
host = os.getenv("HOST").replace("http://", "")  # Rimuove "http://" se presente
dbname = os.getenv("DBNAME")  # DBNAME=AdventureWorksDW

# Costruisco la stringa di connessione
connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"

# Creo l'engine SQLAlchemy
db_engine = sqlalchemy.create_engine(connection_string)

# Eseguo le query SQL sulle due tabelle
dimcustomer = pd.read_sql("SELECT CustomerKey, Title,FirstName, MiddleName, LastName, EmailAddress, Phone, AddressLine1 FROM dimcustomer", db_engine)
```

```python
# TRASFORMIAMO I NOMI DEI CLIENTI IN MODO CHE ABBIANO SOLO LETTERE MINUSCOLE E COGNOMI MAIUSCOLI

import numpy as np
import pandas as pd

dimcustomer['FirstName'] = dimcustomer['FirstName'].str.lower()
dimcustomer['LastName'] = dimcustomer['LastName'].str.upper()
dimcustomer
```

```python
# SULLA EMAIL, UTILIZZANDO IL METODO .STR.SPLIT(), ESTRAIAMO NOME UTENTE E DOMINIO

dimcustomer['NameSplitted'] = dimcustomer['EmailAddress'].str.split(pat="@")
dimcustomer
```

```python
# SULLA COLONNA PHONE, ESTRAIAMO OGNI PARTE DEL NUMERO 

dimcustomer['PhoneSpitted'] = dimcustomer['Phone'].str.split(pat=" ")
dimcustomer
```

```python
# UTILIZZANDO .STR.CONTAINS() , ESTRAIAMO TUTTI GLI INDIRIZZI EMAIL CHE CONTENFONO IL NUMERO 21

dimcustomer.loc[dimcustomer['EmailAddress'].str.contains("21", regex=False)]
```

```python
# ESTRAIAMO TUTTI GLI INDIRIZZI EMAIL CHE CONTENGONO 20 O 10 

dimcustomer_filtrato = dimcustomer.loc[
    dimcustomer['EmailAddress'].str.contains("20|10", regex=True)
]
dimcustomer_filtrato
```

```python
# CALCOLARE LA LUNGHEZZA DI OGNI EMAIL ED ESTRARRE I CINQUE PIU' LUNGHI E I CINQUE PIU' CORTI

dimcustomer['LunghezzaEmail'] = dimcustomer['EmailAddress'].str.len()
dimcustomer
```

```python
# MODIFICARE IL DOMINIO DEGLI INDIRIZZI EMAIL DA "ADVENTURE-WORKS.COM" A "AW-DB.COM" CON .STR.REPLACE()

dimcustomer['EmailAddress'] = dimcustomer['EmailAddress'].str.replace("adventure-works.com", "aw-db.com")
dimcustomer
```

```python
import os
import pandas as pd


cartella = r"C:\Users\Matteo Perrotta\OneDrive\Desktop\archive"

# Lista dei csv
file_list = os.listdir(cartella)

# Lista csv nulli
file_con_nulli = []

# Esaminiamo ogni .csv nella cartella
for nome in file_list:
    if nome.endswith(".csv"):  # Se è un file CSV
        percorso_completo = os.path.join(cartella, nome)
        try:
            # Carichiamo il file in un DataFrame
            df = pd.read_csv(percorso_completo)
            
            # Verifica se ci sono valori NaN nel DataFrame
            num_nulli = df.isna().sum().sum()  # Somma totale dei NaN

            if num_nulli > 0:
                file_con_nulli.append(nome)  # Salviamo il nome del file
        except Exception as e:
            print(f" Errore nel file {nome}: {e}")

# Stampa dei risultati
print("File CSV con dati nulli:")
for nome in file_con_nulli:
    print("-", nome)
```
