```python
#QUALI SONO LE MEDIE DI CHARGES RISPETTO A REGION?
db.groupby('region')['charges'].mean()
```

```python
#QUALI SONO LE MEDIE DI CHARGES RISPETTO A SMOKER?

db['smoker_num'] = db['smoker'].map({'yes': 1, 'no': 0})
db.groupby('region')['smoker_num'].mean()
```

```python
#QUALI SONO LE MEDIE DI CHARGES RISPETTO A SEX?
db['sex_num'] = db['sex'].map({'male': 1, 'female': 0})
db.groupby('region')['sex_num'].mean()
```

```python
#QUALI SONO I DESCRITTORI STATISTICI DI BMI?

descrittori = db['bmi'].describe()
print(descrittori)
```

```python
#QUALI SONO MINIMO, MEDIA E MASSIMO DI CHARGES RISPETTO AI DIVERSI QUARTILI DEI VALORI DI BMI?

db['bmi'].quantile([0.25, 0.5, 0.75])
q1 = db['bmi'].quantile(0.25)
q2 = db['bmi'].quantile(0.5)
q3 = db['bmi'].quantile(0.75)

def bmi_quartile(bmi):
    if bmi <= q1:
        return 'Q1'
    elif bmi <= q2:
        return 'Q2'
    elif bmi <= q3:
        return 'Q3'
    else:
        return 'Q4'

db['bmi_quartile'] = db['bmi'].apply(bmi_quartile)

risultato = db.groupby('bmi_quartile')['charges'].agg(['min', 'mean', 'max'])
print(risultato)
```

```python
#Abbiamo un DataFrame di dipendenti: 
employees_df = pd.DataFrame({ 
    'employee_id': [101, 102, 103, 104, 105], 
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'], 
    'department_id': [1, 2, 1, 2, 3] 
})

employees_df
```

```python
#E un DataFrame di dipartimenti:
departments_df = pd.DataFrame({ 
    'department_id': [1, 2, 3], 
    'department_name': ['HR', 'IT', 'Finance'], 
    'location': ['New York', 'San Francisco', 'Chicago']
})
departments_df
```

```python
# UNIRE QUESTI DATAFRAME IN BASE ALLA COLONNA COMUNE DEPARTMENT_ID IN MODO DA AVERE NEL RISULTATO INFORMAZIONI SIA SUI DIPENDENTI CHE SUI DIPARTIMENTI ('MERGE()')

import pandas as pd
merge = pd.merge(employees_df,departments_df, on='department_id', how='inner')
merge
```

```python
# PER OGNI DATAFRAME TRASFORMA LA COLONNA DEPARTMENT_ID NELL'INDICE, FACENDO IN MODO CHE LA MODIFCIA SIA PERMANENTE; POI UNIRE I DUE DATASET MEDIANTE IL METODO JOIN()

employees_df.set_index("department_id", inplace=True)
departments_df.set_index("department_id", inplace=True)
```

```python
# UNIRE I DATASET MEDIANTE IL METODO JOIN()

merge = employees_df.join(departments_df)
merge
```
